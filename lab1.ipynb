```python
%matplotlib inline
import matplotlib
import numpy as np
```

```python
X = np.arange(0, 10, 0.1)
Y = np.sin(X)
matplotlib.pyplot.plot(X, Y)
```

```python
import math

def R(w):
    num = 10 - 6 * w**2 + 18 * w**2
    denom = (10 - 6 * w**2) ** 2 + 81 * w**2
    return num / denom

def I(w):
    num = 11 * w - 12 * w ** 3
    denom = (10 - 6 * w**2) ** 2 + 81 * w**2
    return num / denom

def H(w):
    r = R(w)
    i = I(w)
    return math.sqrt(r**2 + i**2)

H_Y = np.array(list(map(H, X)))
matplotlib.pyplot.plot(X, H_Y)
```

```python
def F(w):
    r = R(w)
    i = I(w)
    return math.atan2(i, r)

F_Y = np.array(list(map(F, X)))
matplotlib.pyplot.plot(X, F_Y)
```
